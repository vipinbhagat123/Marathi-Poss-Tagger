```python
import sys
import re
import os
from indicnlp.tokenize.sentence_tokenize import sentence_split

# Setup Indic NLP path
INDIC_NLP_LIB_HOME = './indic_nlp_library'
INDIC_NLP_RESOURCES = './indic_nlp_resources'

sys.path.append(INDIC_NLP_LIB_HOME)

from indicnlp import common

# Initialize Indic NLP
common.set_resources_path(INDIC_NLP_RESOURCES)

# ---- Sentence cleaning ----

def is_valid_char(char):
    return (
        '\u0900' <= char <= '\u097F' or  # Devanagari characters
        char.isdigit() or               # English digits
        '\u0966' <= char <= '\u096F' or # Marathi digits (०-९)
        re.match(r"[.,!?;:\"'()\[\]।]", char) or
        char.isspace()
    )

def data_cleaning(sentence):
    """
    Removes non-Marathi characters from the sentence.
    Keeps Devanagari characters, digits, punctuation, and spacing.
    """
    cleaned = ''.join([char for char in sentence if is_valid_char(char)])
    return cleaned.strip()

# ---- Sentence tokenizer ----

def sent_tokenizer(text, lang='mr'):
    """
    Tokenize paragraph into sentences using Indic NLP.
    """
    return sentence_split(text.strip(), lang)

# ---- File processing ----

input_file = "/content/scraped_articles_combined2.1.txt"
output_file = "cleaned_sentences2.1.txt"

with open(input_file, "r", encoding="utf-8") as infile, open(output_file, "w", encoding="utf-8") as outfile:
    for line in infile:
        if line.strip():
            sentences = sent_tokenizer(line, lang='mr')
            for sent in sentences:
                cleaned = data_cleaning(sent)
                if cleaned:
                    outfile.write(cleaned + "\n")

print("✅ Sentences split and cleaned properly. Saved in:", output_file)
```
